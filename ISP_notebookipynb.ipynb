```python
def rmse_euclidean_dist(df_pred, df_test):
    # Calcular la distancia euclidiana
    distance = np.sqrt(np.sum((df_test - df_pred)**2, axis=1))

    # Calcular el error medio
    mean_error = np.mean(distance)

    return mean_error


# Función para aplicar representación positiva
def replace_boundery_value(df):
    # Encuentra el valor mínimo en el DataFrame
    min_value = df.min().min()
    df.replace(100, min_value, inplace=True)

    return df


scorer = make_scorer(rmse_euclidean_dist, greater_is_better=False)
```

```python
X_train = replace_boundery_value(DSI1_trnrss)
X_test = replace_boundery_value(DSI1_tstrss)
y_train = DSI1_trncrd
y_test = DSI1_tstcrd

# Definir la función de puntuación utilizando la métrica RMSE
scorer = make_scorer(rmse_euclidean_dist, greater_is_better=False)
start_time = time.time()
# Definir el rango de valores de k que deseas probar
k_values = list(range(1, 52))

# Lista para almacenar los resultados de la validación cruzada para cada valor de k
cross_val_scores = []

# Realizar la validación cruzada para cada valor de k
for k in k_values:
    knn = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=10, scoring=scorer)
    cross_val_scores.append(np.mean(scores))
# Encontrar el valor óptimo de k que minimiza el error
optimal_k = k_values[np.argmax(cross_val_scores)]

print(f"El mejor valor para k es: {optimal_k}")

start_time_1_model = time.time()
# Entrenar el modelo final con el valor óptimo de k en el conjunto de entrenamiento
knn_final_model_DSI1 = KNeighborsRegressor(n_neighbors=optimal_k)
knn_final_model_DSI1.fit(X_train, y_train)
# Realizar predicciones en el conjunto de prueba
y_knn_pred_DSI1 = knn_final_model_DSI1.predict(X_test)
end_time = time.time()

elapsed_time = end_time - start_time
elapsed_time_1_model = end_time - start_time_1_model
print(f"Total elapsed_time: {elapsed_time}")
print(f"Elapsed_time optimal model: {elapsed_time_1_model}")
# Calcular el RMSE en el conjunto de prueba
knn_rmse_test_DSI1 = rmse_euclidean_dist(y_test[['x', 'y','z']], y_knn_pred_DSI1[:,:3])
print(f"RMSE en el conjunto de DSI1 prueba con k óptimo (KNN): {knn_rmse_test_DSI1}")
```

```python
RF = RandomForestRegressor(random_state=42)

def RF_param_search(X, y, nfolds):
    # Number of trees in random forest
    n_estimators =  [10, 20, 30, 40, 50, 60, 70, 80, 90]
    # Number of features to consider at every split
    max_features = ['sqrt', 'log2',None]
    # Maximum number of levels in tree
    max_depth = [10, 20, 30, 40, 50, 70, 90, None]
    min_samples_split = [2, 5, 10]
    min_samples_leaf = [1, 2, 3, 4]

    # Create the random grid    
    random_grid = {'n_estimators': n_estimators,
                   'max_features': max_features,
                   'max_depth': max_depth,
                   'min_samples_split': min_samples_split,
                   'min_samples_leaf': min_samples_leaf,
                  }
    
    rf_random = RandomizedSearchCV(estimator = RF, param_distributions = random_grid, n_iter = 300, cv = nfolds, n_jobs = -1)
    rf_random.fit(X, y)
    rf_random.best_params_
    pprint(rf_random.best_params_)
    
    return rf_random.best_params_

def get_grid_n_estimators(input_num):

    input_num = max(input_num, 10)

    # Calculate the starting point for the list
    start_point = input_num - 10

    # Generate a list of 6 integers excluding 0
    result_list = [max(start_point + i * 10, 10) for i in range(1,7)]

    return result_list


def get_grid_max_depth(input_num):
    input_num = max(input_num, 10)

    # Calculate the starting point for the list
    start_point = input_num - 10

    # Generate a list of 6 integers excluding 0
    result_list = [max(start_point + i * 10, 10) for i in range(1,6)]

    return result_list

def get_grid_min_samples(input_num):
    
    input_num = max(input_num, 1)

    # Calculate the starting point for the list
    if ((input_num - 3)<0): 
        start_point = 1
    else:
        start_point = input_num - 3

    # Generate a list of 11 integers excluding 0 with an interval of 1
    result_list = [max(start_point + i *1, 2) for i in range(1, 5)]
    
    return result_list
    
def RF_grid_param_search(X, y, estimator, depth, split, leaf, nfolds):
    # Number of trees in random forest
    n_estimators =  get_grid_n_estimators(estimator)
    # Number of features to consider at every split
    max_features = ['sqrt', 'log2',None]
    # Maximum number of levels in tree
    max_depth = get_grid_max_depth(depth)
    min_samples_split = get_grid_min_samples(split)
    min_samples_leaf = get_grid_min_samples(leaf)

    # Create the random grid    
    param_grid = {'n_estimators': n_estimators,
                   'max_features': max_features,
                   'max_depth': max_depth,
                   'min_samples_split': min_samples_split,
                   'min_samples_leaf': min_samples_leaf,
                  }

    rf_random = GridSearchCV(estimator = RF, param_grid = param_grid, cv = nfolds, n_jobs = -1)
    rf_random.fit(X, y)
    rf_random.best_params_
    pprint(rf_random.best_params_)
    
    return rf_random.best_params_
```

# SVR

```python
def svr_param_selection(X, y, d, nfolds):
    C = [0.01, 0.1, 1, 5, 8]
    epsilon = [0.1, 0.2, 0.5, 0.8, 1]
    param_grid = {'C': C,  'epsilon': epsilon}
    grid_search = GridSearchCV(SVR(kernel='rbf'), param_grid, cv=nfolds)
    grid_search.fit(X, np.array(y[d]))
    grid_search.best_params_
    return grid_search.best_params_
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train = replace_boundery_value(DSI1_trnrss)
X_test = replace_boundery_value(DSI1_tstrss)
y_train = DSI1_trncrd
y_test = DSI1_tstcrd

scaler = MaxAbsScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

start_time = time.time()

best_x = svr_param_selection(X_train_scaled,y_train, 'x', 5)
best_y = svr_param_selection(X_train_scaled,y_train, 'y', 5)
best_z = svr_param_selection(X_train_scaled,y_train, 'z', 5)

pprint(best_x)
pprint(best_y)
pprint(best_z)

start_time_opt_model = time.time()

# Entrenar modelos para cada dimensión
svm_regressor_x_DSI1 = SVR(kernel='rbf', C=best_x["C"], epsilon=best_x["epsilon"])
svm_regressor_y_DSI1 = SVR(kernel='rbf', C=best_y["C"], epsilon=best_y["epsilon"])
svm_regressor_z_DSI1 = SVR(kernel='rbf', C=best_z["C"], epsilon=best_z["epsilon"])
#svm_regressor_y_DSI1 = SVR(kernel='rbf', C=10, gamma=1, epsilon=0.5)
#svm_regressor_z_DSI1 = SVR(kernel='rbf', C=0.001, gamma=0.001, epsilon=0.5)

# Entrenar los modelos para cada dimensión
svm_regressor_x_DSI1.fit(X_train_scaled, y_train.iloc[:, 0])  # 0 representa la columna x
svm_regressor_y_DSI1.fit(X_train_scaled, y_train.iloc[:, 1])  # 1 representa la columna y
svm_regressor_z_DSI1.fit(X_train_scaled, y_train.iloc[:, 2])  # 2 representa la columna z

# Realizar predicciones en el conjunto de prueba
y_pred_x_DSI1 = svm_regressor_x_DSI1.predict(X_test_scaled)
y_pred_y_DSI1 = svm_regressor_y_DSI1.predict(X_test_scaled)
y_pred_z_DSI1 = svm_regressor_z_DSI1.predict(X_test_scaled)

end_time = time.time()

elapsed_time = end_time - start_time
elapsed_time_opt_model = end_time - start_time_opt_model

print(f"Total elapsed_time: {elapsed_time}")
print(f"Elapsed_time optimal model: {elapsed_time_1_model}")

# Combina las predicciones en un array bidimensional
y_pred_combined_DSI1 = np.column_stack((y_pred_x_DSI1, y_pred_y_DSI1, y_pred_z_DSI1))

# Evaluar el rendimiento del modelo
rmse_DSI1 = rmse_euclidean_dist(y_test[['x', 'y','z']], y_pred_combined_DSI1)

print(f"Raíz del error medio cuadrático (RMSE) mediante SVR en DSI1: {rmse_DSI1}")
```

```python
import warnings

# Set warnings to be ignored
warnings.filterwarnings("ignore")

def prediction_by_row(model, X):

    # Initialize an empty list to store predictions
    predictions = []

    # Iterate through rows in the test data
    for index, row in X.iterrows():
        # Convert the row to a NumPy array and reshape
        row_array = row.to_numpy().reshape(1, -1)

        # Make predictions using the model and append to the list
        y_pred_row = model.predict(row_array)
        predictions.append(y_pred_row)

    # Convert the list of predictions to a NumPy array
    predictions_array = np.array(predictions)

    predictions_array_2d = predictions_array.reshape(len(X), 5)

    # Convert the NumPy array to a DataFrame
    predictions_df = pd.DataFrame(predictions_array_2d, columns=['x', 'y', 'z', 'floor', 'building'])

    return predictions_df

def convert_floor_to_z(floor):
    return round(floor*3.7,1)
```
